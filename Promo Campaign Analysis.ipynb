```python
from scipy.stats import shapiro
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Pilih kolom-kolom yang ingin diuji
kolom_uji = ['Frequency_Score', 'Monetary_Score', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5','Response']

# Buat dataframe untuk menyimpan hasil uji
hasil_normalitas = {}

for col in kolom_uji:
    stat, p_value = shapiro(df[col])
    hasil_normalitas[col] = {
        'W-Statistic': stat,
        'p-value': p_value,
        'Normal': 'Yes' if p_value > 0.05 else 'No'
    }

normalitas_df = pd.DataFrame(hasil_normalitas).T
```

```python
# Pastikan p-value bertipe float
normalitas_df['p-value'] = normalitas_df['p-value'].astype(float)

# Visualisasi heatmap p-value
plt.figure(figsize=(8, 2))
sns.heatmap(normalitas_df[['p-value']].T, annot=True, cmap='RdYlGn', cbar=False, fmt=".3f")
plt.title('Heatmap Uji Normalitas (Shapiro-Wilk Test)')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Kolom Campaign
campaign_cols = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']

# Kolom LRFM
lrfm_cols = ['Frequency', 'Monetary']  # pastikan ini sesuai dengan kolom di df

# Gabungkan untuk korelasi
cols_to_corr = campaign_cols + lrfm_cols
corr_data = df[cols_to_corr]


# Korelasi Spearman
spearman_corr = corr_data.corr(method='spearman')


import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(
    spearman_corr.loc[campaign_cols, lrfm_cols],
    annot=True,
    cmap='coolwarm',
    fmt=".2f"
)
plt.title('Spearman Correlation: Campaign Acceptance vs LRFM')
plt.xlabel('LRFM Features')
plt.ylabel('Campaign Response')
plt.tight_layout()
plt.show()
```

```python
# Buat salinan data dan reshape untuk visualisasi
df_melted = df.melt(
    id_vars='Segment',
    value_vars=['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response'],
    var_name='campaign',
    value_name='accepted'
)

# Hitung rata-rata accepted per campaign per segment
campaign_by_segment = df_melted.groupby(['Segment', 'campaign'])['accepted'].mean().reset_index()
```

```python
pivot_table = campaign_by_segment.pivot(index='Segment', columns='campaign', values='accepted')

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Proporsi Accepted Campaign per Segmen')
plt.ylabel('Segment')
plt.xlabel('Campaign')
plt.tight_layout()
plt.show()
```
